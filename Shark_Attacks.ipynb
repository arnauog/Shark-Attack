```python
import chardet

# Read the file in binary mode to detect encoding
with open('attacks.csv', 'rb') as f:
    raw_data = f.read()
    result = chardet.detect(raw_data)
    encoding = result['encoding']
    print(f"Detected encoding: {encoding}")
```
